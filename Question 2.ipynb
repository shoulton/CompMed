```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The location of each pixel
nX = 200
nY = 200
# Create a blank image
I = np.zeros((nX, nY))
cx = 75 # x component of center
cy = 75 # y component of center
rx = 30 # width of half rectangle
ry = 45 # height of half rectangle

for i in range(nX):
    for j in range(nY):
        if ((i < (cx + rx)) and (i > (cx - rx)) and (j < (cy + ry)) and (j > (cy - ry))):
            I[j,i] = 100
            
# Display
plt.figure(figsize=(10,10))
plt.imshow(I, origin='lower') #put origin at lower left
```

```python
# initialize an image of all zeros
ITransformed = np.zeros((nY, nX))
for i in range(nX): # loop through each row
    for j in range(nY): # loop through each column
        # we are looking for the value to assign to Ibx[j,i]        
        
        # find the position to look at in the image J
        iLook = i * S # you should implement this line
        jLook = j * S # you should implement this line
        
        # round them to the nearest integer
        iLookRound = int(np.round(iLook))
        jLookRound = int(np.round(jLook))
    
        
        if ((iLookRound < 1) or (iLookRound > nX) or (jLookRound < 1) or (jLookRound > nY)):
            # if so, fill the image with the value zero
            ITransformed[j, i] = 0
        else:
            # otherwise, assign the value in our image at this point
            ITransformed[j][i] = I[jLookRound][iLookRound]

# don't forget to index your images by [row,column] and not [x,y]

            
# Display
plt.figure(figsize=(10,10))
plt.imshow(ITransformed, origin='lower') #put origin at lower left
```

The image shrunk by a factor of S, rather than growing to match J. 

```python
# initialize an image of all zeros
ITransformed = np.zeros((nY, nX))
for i in range(nX): # loop through each row
    for j in range(nY): # loop through each column
        # we are looking for the value to assign to Ibx[j,i]        
        
        # find the position to look at in the image J
        iLook = i / S # you should implement this line
        jLook = j / S # you should implement this line
        
        # round them to the nearest integer
        iLookRound = int(np.round(iLook))
        jLookRound = int(np.round(jLook))
    
        
        if ((iLookRound < 1) or (iLookRound > nX) or (jLookRound < 1) or (jLookRound > nY)):
            # if so, fill the image with the value zero
            ITransformed[j, i] = 0
        else:
            # otherwise, assign the value in our image at this point
            ITransformed[j][i] = I[jLookRound][iLookRound]

# don't forget to index your images by [row,column] and not [x,y]

            
# Display
plt.figure(figsize=(10,10))
plt.imshow(ITransformed, origin='lower') #put origin at lower left
```

The image grew, and now matches J. It functions differently because we are transforming points in the new image to match the old image, and so whatever we do has the inverse effect. By inverting S, we get the scaling we expected.
